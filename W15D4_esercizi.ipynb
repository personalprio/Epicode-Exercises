### Esercizio 3
Abbiamo il seguente array NumPy: linear_data = np.array([x for x in range(27)]) Lo ridimensioniamo mediante il metodo .reshape(): reshaped_data = linear_data.reshape((3, 3, 3)) Quante dimensioni ha il nuovo array? Come facciamo per accedere ai singoli elementi?

```python
#Divido l'esercizio in due celle così da mostrare le differenze
linear_data = np.array([x for x in range(27)])
print(linear_data)
```

```python
reshaped_data = linear_data.reshape((3,3,3))
#Abbiamo ora un array tridimensionale, lo immagino come un cubo ma potrei sbagliarmi
print(reshaped_data)
#Per accedere sempre all'elemento di valore 2, negli indico scrivo prima il sottoinsieme, poi la riga, poi la colonna
print(reshaped_data[0,0,2])
```
